# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
df.Area = label.fit_transform(df['Area'])
```

```python
 df.Area.head()
```

```python
import numpy as np
```

```python
for column in df.columns:
    if df[column].dtype == np.number:
        continue 
    df[column] = label.fit_transform(df[column])
```

```python
df.head()
```

```python
#another way
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in df.columns:
    if is_numeric_dtype(df[column]):
        continue
    else:
        df[column] = LabelEncoder().fit_transform(df[column])
```

# Ordinal Encoder

```python
df = pd.read_csv('supershop.csv')
df.head()
```

```python
df.Area.unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder 
```

```python
city = ['Dhaka', 'Ctg', 'Rangpur']
ordinal = OrdinalEncoder(categories=[city])
encoded = ordinal.fit_transform(df[['Area']])
```

```python
encoded
```

```python
encoded_area = pd.DataFrame(encoded,columns=['Encoded'])
```

```python
df.Area = encoded_area
```

```python
df.Area
```

```python
df = pd.read_csv('supershop.csv')
df.head()
```

```python
cols=['Area']
for col in cols:
    uniques = df[col].unique()
    df[col] = OrdinalEncoder(categories=[uniques]).fit_transform(df[[col]])
```
